## 실습 2: aromawine 데이터로 ridge 회귀 진행하고, mse 값과 R^2 구하기 (alpha값 0.001, 0.01, 0.1)

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
wine = pd.read_excel('wine_aroma.xlsx')
wine.head()
```

```python
Input = wine.drop(columns=['Aroma'])
Output = wine['Aroma']
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=0.001)
ridge.fit(Input, Output)

output_pred = ridge.predict(Input)
print('R^2 : ', r2_score(Output, output_pred))
print('MSE : ', mean_squared_error(Output, output_pred))
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=0.01)
ridge.fit(Input, Output)

output_pred = ridge.predict(Input)
print('R^2 : ', r2_score(Output, output_pred))
print('MSE : ', mean_squared_error(Output, output_pred))
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=0.1)
ridge.fit(Input, Output)

output_pred = ridge.predict(Input)
print('R^2 : ', r2_score(Output, output_pred))
print('MSE : ', mean_squared_error(Output, output_pred))
```

alpha값이 작을 수록 mse 값이 작아지는 것을 알 수 있다.  
alpha값이 작을 수록 r_square가 1에 가까워지는 것을 알 수 있다.
